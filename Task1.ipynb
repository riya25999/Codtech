```python
# Install necessary libraries
!pip install pandas numpy matplotlib seaborn

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Download the dataset (Iris dataset from UCI Repository)
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data -O iris.csv

# Define column names (Iris dataset doesn't come with headers)
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

# Load the dataset
df = pd.read_csv("iris.csv", header=None, names=columns)

# Display the first few rows
df.head()
```

```python
# Display the shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Check for null values
print("\nNull values in the dataset:")
print(df.isnull().sum())

# Check data types of each column
print("\nData types of columns:")
print(df.dtypes)

# Summary statistics for numeric columns
print("\nSummary statistics:")
print(df.describe())
```

```python
# Plot histograms for all numeric features
df.hist(figsize=(10, 8), bins=20, color='lightblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
# Compute correlation matrix for numeric columns
correlation_matrix = df.iloc[:, :-1].corr()  # Exclude species column (categorical)

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Visualize pairwise relationships in the dataset
sns.pairplot(df, hue="species", diag_kind="kde")
plt.suptitle("Pairplot of Features by Species", y=1.02)
plt.show()
```

```python
# Scatter plot between sepal length and petal length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species", palette="viridis", alpha=0.8)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.show()
```

```python
# Box plot for sepal length
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="species", y="sepal_length", palette="Set2")
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```
